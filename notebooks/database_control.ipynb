```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from pandasql import sqldf

from wordcloud import WordCloud

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression

import plotly.express as px

sns.set_style("whitegrid")

print("All libraries loaded successfully")
```

```python
df = pd.read_csv("../data/master_dataset.csv")

print("Dataset loaded")

print("Shape:", df.shape)

df.head()
```

```python
df.info()

df.describe()
```

```python
print("Total comments:", len(df))
```

```python
df["comment_length"] = df["comment"].astype(str).apply(len)

df.head()
```

```python
plt.figure(figsize=(10,5))

df["video_id"].value_counts().plot(kind="bar")

plt.title("Comments per Video")

plt.xlabel("Video ID")
plt.ylabel("Number of Comments")

plt.show()
```

Visualization: Comment length distribution

```python
plt.figure(figsize=(10,5))

sns.histplot(df["comment_length"], bins=50)

plt.title("Comment Length Distribution")

plt.show()
```

```python
text = " ".join(df["comment"].astype(str))

wordcloud = WordCloud(width=800, height=400).generate(text)

plt.figure(figsize=(12,6))

plt.imshow(wordcloud)

plt.axis("off")

plt.title("Most Frequent Words")

plt.show()
```

```python
vectorizer = CountVectorizer(stop_words="english")

X = vectorizer.fit_transform(df["comment"].astype(str))

word_counts = pd.DataFrame({
    "word": vectorizer.get_feature_names_out(),
    "count": X.sum(axis=0).A1
})

word_counts.sort_values("count", ascending=False).head(20)
```

```python
positive_words = ["good", "great", "excellent", "amazing", "love", "best"]
negative_words = ["bad", "worst", "poor", "hate", "terrible"]

def get_sentiment(comment):

    comment = str(comment).lower()

    if any(word in comment for word in positive_words):
        return "positive"

    elif any(word in comment for word in negative_words):
        return "negative"

    else:
        return "neutral"


df["sentiment"] = df["comment"].apply(get_sentiment)

df.head()
```

```python
sns.countplot(data=df, x="sentiment")

plt.title("Sentiment Distribution")

plt.show()
```

```python
print("===== DASHBOARD SUMMARY =====")

print("Total Comments:", len(df))

print("\nComments per video:")
print(df["video_id"].value_counts())

print("\nSentiment Distribution:")
print(df["sentiment"].value_counts())
```

```python
df.to_csv("../data/analyzed_comments.csv", index=False)

print("Analysis exported successfully")
```
